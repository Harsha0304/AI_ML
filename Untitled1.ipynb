```python
#scatterplot
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5,6,8])
y=np.array([2,3,4,5,6,7,8])
size=np.array([10,20,30,40,50,60,70])
plt.scatter(x,y , s=size, alpha=0.2)
plt.show
```

```python
#scatterplot random array and colormap
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randint(50,size=(50))
y=np.random.randint(50,size=(50))
size=10*np.random.randint(50,size=(50))
color=np.random.randint(50,size=(50))
plt.scatter(x,y , s=size,c=color,alpha=0.8,cmap='gist_rainbow')
plt.colorbar()
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

x=np.array([22,33,4,22,45,52,4,56,44,33])
y=np.array([10,90,80,70,60,50,40,30,20,10])

s,i,r,p,std_error=stats.linregress(x,y)

def my_fun(x):
    l=s*x+i
    return l

y_new= list(map(my_fun,x))
plt.scatter(x,y)
plt.plot(x,y_new)
plt.show()
```
